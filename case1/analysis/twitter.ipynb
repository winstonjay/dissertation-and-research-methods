```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import json
import datetime

def load_file(fname, mapfile):
    m = pd.read_csv(mapfile)
    relevent = set(m.idx)
    data = []
    with open(fname) as fp:
        for i, line in enumerate(fp):
            if i not in relevent:
                continue
            data.append(json.loads(line))
    return pd.DataFrame(data)

def parse_tw_url(a):
    page_url = a.get('url', None)
    urls = []
    if page_url:
        urls.append(urlex(page_url['urls'][0]))
    desc_urls = a['description']['urls']
    if desc_urls:
        for u in desc_urls:
            urls.append(urlex(u))
    return set(urls)

def urlex(u):
    url = u.get("expanded_url", None)
    if url is None or url == 'null':
        url = u.get('url', None)
    return url.lower()

tw_metoo  = load_file('../data/twitter/metoo/metoo.ndjson',  '../data/twitter/metoo/metoo.map.csv')
tw_blm    = load_file('../data/twitter/blm/blm.ndjson',      '../data/twitter/blm/blm.map.csv')
tw_occupy = load_file('../data/twitter/occupy/occupy.ndjson', '../data/twitter/occupy/occupy.map.csv')

tw_metoo.created_at  = pd.to_datetime(tw_metoo.created_at, format='%a %b %d %H:%M:%S +0000 %Y')
tw_blm.created_at    = pd.to_datetime(tw_blm.created_at, format='%a %b %d %H:%M:%S +0000 %Y')
tw_occupy.created_at = pd.to_datetime(tw_occupy.created_at, format='%a %b %d %H:%M:%S +0000 %Y')

tw_metoo.urls  = tw_metoo.entities.apply(parse_tw_url)
tw_blm.urls    = tw_blm.entities.apply(parse_tw_url)
tw_occupy.urls = tw_occupy.entities.apply(parse_tw_url)
```

```python
now = datetime.datetime.now()
tw_metoo.days  = (now - tw_metoo.created_at).dt.days
tw_blm.days    = (now - tw_blm.created_at).dt.days
tw_occupy.days = (now - tw_occupy.created_at).dt.days
```
